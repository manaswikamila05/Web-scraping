# Filtering jobs posted today based on a skill

```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import time
```

```python
url = "https://www.timesjobs.com/candidate/job-search.html?searchType=Home_Search&from=submit&asKey=OFF&txtKeywords=&cboPresFuncArea=35"
```

```python
def find_jobs():
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')

    jobs = soup.find_all('li', class_ = "clearfix job-bx wht-shd-bx")

    for index, job in enumerate(jobs):
        published_date = job.find('span', class_= "sim-posted").span.text

        if 'today' in published_date:
            company_name = job.find('h3', class_ = "joblist-comp-name").text.strip().split()[0]
            skills = job.find('span', class_ = "srp-skills").text.strip().replace(' ', "")

            more_info = job.header.h2.a['href']

            if familar_skill in skills:
                print(f"Company Name: {company_name}")
                print(f"Required skills: {skills.strip()}")
                print(f"More Info: {more_info}")
                print()
                
```
